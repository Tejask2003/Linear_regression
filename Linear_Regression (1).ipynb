```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for demonstration
np.random.seed(0)
X = np.random.rand(100, 1) * 20  # Generating random feature between 0 and 20
y = 3 * X + 5 + np.random.randn(100, 1) * 5  # Linear relationship with some noise

# Implementing Linear Regression from scratch
class LinearRegression:
    def init(self):
        self.theta = None  # Parameters (weights)

    def fit(self, X, y):
        # Add intercept term
        X = np.hstack((np.ones((X.shape[0], 1)), X))
        # Calculate parameters using Normal Equation
        self.theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    def predict(self, X):
        # Add intercept term
        X = np.hstack((np.ones((X.shape[0], 1)), X))
        return X.dot(self.theta)

# Splitting data into training and testing sets (for demonstration)
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

# Instantiate and fit the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
predictions_train = model.predict(X_train)
predictions_test = model.predict(X_test)

# Plotting the data and the regression line for both training and testing datasets
plt.figure(figsize=(10, 5))

# Plot for training dataset
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, label='Training data', color='orange', marker='o', alpha=0.7)
plt.plot(X_train, predictions_train, color='blue', linestyle='-', linewidth=2, label='Linear regression')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression - Training Dataset (Modified)')
plt.legend()

# Plot for testing dataset
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, label='Testing data', color='green', marker='x', alpha=0.7)
plt.plot(X_test, predictions_test, color='red', linestyle='--', linewidth=2, label='Linear regression')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression - Testing Dataset (Modified)')
plt.legend()

plt.tight_layout()
plt.show()
```
